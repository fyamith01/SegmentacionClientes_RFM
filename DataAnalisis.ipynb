#### **1.1.2 Formato de fecha adecuado**
Ahora debemos garantizar que el formato de las fechas sea correcto, para esto, nos aseguramos de que todos los registros de la columna ```fecha_efectiva``` tengan 8 caracteres, de lo contrario, quiere deccir que no está la fecha completa, y para este ejercicio, vamos a ignorar estos valores incompletos, por simplicidad. (Dado que hay varios registros que solo tienen formato de año, como 2020, pero no dicen nada más). Para los valores que tienen 6 caracteres vamos a hacer un tratamiento especial, aquellos que son del tipo '202012' los vamos a tomar como si fuera el primer día de ese mes, ya que aunque se podría ignorar esa fecha por estar incompleta, ella regsitra una transacción real y en un mes conocido, entonces si cae en el intervalo de los últimos 6 meses es información útil.

````python
def format_fecha(x):
    """
    Formatea la fecha según su longitud:
    - 4 caracteres o un número distinto de 6 o 8: nulo
    - 6 caracteres: agrega '01' al final (primer día del mes)
    - 8 caracteres: mantiene como está
    - otros: nulo
    """
    if not isinstance(x, str):
        return None
    
    x = x.strip()
    
    if len(x) == 6:
        return x + '01'
    elif len(x) == 8:
        return x
    else:
        return None
````

````python
# Aplicar la función de formateo
df_cliente['fecha_efectiva'] = df_cliente['fecha_efectiva'].apply(format_fecha)
````

#### **1.1.9 Feature Selection - Selección de variables clave para agrupar**

Aquí vamos a empezar a hacer todas las agrupaciones necesarias para obtener la información de cada cliente. acerca de su última transacción, la frecuencia entre transacciones y el valor promedio de dichas transacciones, en el rango de los últimos 6 meses. 

````python
# Agregación principal
base = df_analisis.groupby('Id_cliente').agg(
    last_date=('fecha_efectiva', 'max'),
    frequency_total=('Id_tx', 'count'),
    monetary_total=('valor', 'sum'),
    monetary_avg=('valor', 'mean'),
    monetary_std=('valor', 'std')
)

base['frequency_total'] = base['frequency_total'].fillna(0)
base['monetary_total'] = base['monetary_total'].fillna(0)
base['monetary_avg'] = base['monetary_avg'].fillna(0)
base['monetary_std'] = base['monetary_std'].fillna(0)

base.head()
````

Empezamos calculando las columnas base para análisis de clientes. Agrupamos la frecuencia de transacción por cliente, así como la suma de todos los valores del total de sus transacciones, el promedio de todos los valores del total de sus transacciones, la desviación estándar de todos los valores del total de sus transacciones y la última fecha en la que realizó una transacción cada cliente, esta última sirve para hallar posteriormente la recencia o ***recency***

````python
# Calcular recency
base['recency'] = (ultima_fecha - base['last_date']).dt.days.fillna(999)
base
````

Con esto ya tenemos la columna que nos da información de la recencia, nos dice hace cuántos días un cliente realizó la última transacción, y si nunca ha realizado una, aparece un 999, haciendo referencia a que recientemente no ha hecho transacciones, sino que la última fue en un tiempo demasiado lejano

````python
# Para métricas por tipo de transacción (si necesitas mantener esa lógica)
for clase_valor in ['COMPRA', 'AVANCE', 'SEGURO']:
    clase_data = df_analisis[df_analisis['clase'] == clase_valor].groupby('Id_cliente').agg(
        frequency=('Id_tx', 'count'),
        monetary=('valor', 'sum'),
        last_date=('fecha_efectiva', 'max')
    ).add_prefix(f'{clase_valor.lower()}_')
    
    base = base.join(clase_data, how='left')

    # rellenar nulos con 0 para frequency y monetary
    base[f'{clase_valor.lower()}_frequency'] = base[f'{clase_valor.lower()}_frequency'].fillna(0)
    base[f'{clase_valor.lower()}_monetary'] = base[f'{clase_valor.lower()}_monetary'].fillna(0)


base.head()
````

Calculamos ahora columnas de frecuencia, valor promedio y fecha de última transacción, por cliente para cada clase. Esto nos provee de información valiosa y más específica acerca de las transacciones por cliente. Es decir, ahora agrupamos por cliente la información específica de las transacciones por clase.

````python
#calcular recency por clase
for clase_valor in ['COMPRA', 'AVANCE', 'SEGURO']:
    base[f'{clase_valor.lower()}_recency'] = (ultima_fecha - base[f'{clase_valor.lower()}_last_date']).dt.days.fillna(999)

print("Tabla agregada por cliente:")
base.head()
````

Finalmente, nos aseguramos de tener la variable de recency para cada clase, y por supuesto, con el manejo adecuado para los valores nulos. Si un cliente nunca ha hecho una transacción de determinada clase, entonces la columna ```{clase}_recency``` tomará el valor 999 en dicha posición, acorde a lo establecido previamente, ya que estos nulos no pueden tomar el valor 0, porque esto significaría que recientemente el cliente realizó una transacción, no solo recientemente, hoy mismo.

````python
# selección de columnas finales
rfm_clases = base[[
    'recency', 'frequency_total', 'monetary_total',
    'compra_frequency', 'compra_monetary', 'compra_recency',
    'avance_frequency', 'avance_monetary', 'avance_recency',
    'seguro_frequency', 'seguro_monetary', 'seguro_recency'
]]

rfm_clases
````

#### **1.2.2 Aplicando Algoritmo de Clustering K-Means para segmentar**

Primero usaremos la siguiente función para hallar el número óptimo de clusters

````python
# función para encontrar número óptimo de clusters, pervio al algoritmo principal
def encontrar_numero_optimo_clusters(df, max_clusters=10, metodo='todos'):
    """
    Encuentra el número óptimo de clusters usando múltiples métodos
    
    Parámetros:
    -----------
    df : pandas DataFrame
        DataFrame con las features escaladas
    max_clusters : int
        Número máximo de clusters a evaluar
    metodo : str
        'elbow' : solo método del codo
        'silhouette' : solo silhouette
        'todos' : todos los métodos
    
    Returns:
    --------
    dict : Diccionario con resultados y número óptimo sugerido
    """
    
    # Preparar datos
    X = df.values
    
    # Rango de clusters a evaluar
    k_range = range(2, max_clusters + 1)
    
    # Métricas a calcular
    inertia = []
    silhouette_scores = []
    calinski_scores = []
    davies_scores = []
    
    print("=" * 70)
    print("ANÁLISIS DEL NÚMERO ÓPTIMO DE CLUSTERS")
    print("=" * 70)
    
    # Calcular métricas para cada k
    for k in k_range:
        print(f"Evaluando k = {k}...", end="\r")
        
        # KMeans clustering
        kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
        labels = kmeans.fit_predict(X)
        
        # Calcular métricas
        inertia.append(kmeans.inertia_)
        
        if len(set(labels)) > 1:  # Necesario para silhouette y otras métricas
            silhouette_scores.append(silhouette_score(X, labels))
            calinski_scores.append(calinski_harabasz_score(X, labels))
            davies_scores.append(davies_bouldin_score(X, labels))
        else:
            silhouette_scores.append(0)
            calinski_scores.append(0)
            davies_scores.append(float('inf'))
    
    print("\n" + "=" * 70)
    
    # Determinar número óptimo según cada método
    resultados = {}
    
    # 1. Método del Codo
    if metodo in ['elbow', 'todos']:
        # Calcular diferencias
        diff = np.diff(inertia)
        diff2 = np.diff(diff)
        
        # Método 1: Punto donde la reducción de inercia cambia significativamente
        # Normalizar diferencias
        diff_normalized = diff / inertia[1:]
        elbow_point = None
        
        # Buscar punto donde la reducción relativa es menor que un umbral
        for i in range(len(diff_normalized)):
            if i > 0 and diff_normalized[i] < 0.1 * diff_normalized[i-1]:
                elbow_point = k_range[i]
                break
        
        # Método 2: Punto de máxima curvatura
        if elbow_point is None:
            # Calcular ángulos
            angles = []
            for i in range(1, len(inertia)-1):
                # Vectores
                v1 = np.array([k_range[i] - k_range[i-1], inertia[i] - inertia[i-1]])
                v2 = np.array([k_range[i+1] - k_range[i], inertia[i+1] - inertia[i]])
                # Ángulo entre vectores
                cos_angle = np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))
                angle = np.arccos(np.clip(cos_angle, -1, 1))
                angles.append(angle)
            
            if angles:
                elbow_point = k_range[np.argmax(angles) + 1]
        
        resultados['elbow'] = elbow_point if elbow_point else k_range[0]
    
    # 2. Método de Silhouette
    if metodo in ['silhouette', 'todos']:
        if silhouette_scores:
            silhouette_optimo = k_range[np.argmax(silhouette_scores)]
            resultados['silhouette'] = silhouette_optimo
    
    # 3. Método de Calinski-Harabasz (mayor es mejor)
    if metodo == 'todos':
        if calinski_scores:
            calinski_optimo = k_range[np.argmax(calinski_scores)]
            resultados['calinski'] = calinski_optimo
    
    # 4. Método de Davies-Bouldin (menor es mejor)
    if metodo == 'todos':
        if davies_scores:
            davies_optimo = k_range[np.argmin(davies_scores)]
            resultados['davies'] = davies_optimo
    
    # Sugerencia final
    if metodo == 'todos':
        # Promedio de sugerencias (excluyendo None)
        sugerencias = [v for v in resultados.values() if v is not None]
        if sugerencias:
            sugerencia_final = int(np.round(np.mean(sugerencias)))
        else:
            sugerencia_final = 3
    else:
        sugerencia_final = list(resultados.values())[0]
    
    resultados['sugerencia_final'] = sugerencia_final
    
    # Visualizaciones
    fig, axes = plt.subplots(2, 3, figsize=(18, 10))
    
    # Gráfico 1: Método del Codo
    axes[0, 0].plot(k_range, inertia, 'bo-', linewidth=2, markersize=8)
    axes[0, 0].set_xlabel('Número de Clusters (k)', fontsize=12)
    axes[0, 0].set_ylabel('Inercia (Within-Cluster Sum of Squares)', fontsize=12)
    axes[0, 0].set_title('Método del Codo', fontsize=14, fontweight='bold')
    axes[0, 0].grid(True, alpha=0.3)
    
    # Marcar punto del codo si existe
    if 'elbow' in resultados and resultados['elbow']:
        axes[0, 0].axvline(x=resultados['elbow'], color='red', linestyle='--', 
                          label=f'Codo sugerido: k={resultados["elbow"]}')
        axes[0, 0].legend()
    
    # Gráfico 2: Silhouette Score
    if silhouette_scores:
        axes[0, 1].plot(k_range, silhouette_scores, 'go-', linewidth=2, markersize=8)
        axes[0, 1].set_xlabel('Número de Clusters (k)', fontsize=12)
        axes[0, 1].set_ylabel('Silhouette Score', fontsize=12)
        axes[0, 1].set_title('Método de Silhouette', fontsize=14, fontweight='bold')
        axes[0, 1].grid(True, alpha=0.3)
        
        if 'silhouette' in resultados:
            axes[0, 1].axvline(x=resultados['silhouette'], color='red', linestyle='--',
                              label=f'Óptimo: k={resultados["silhouette"]}')
            axes[0, 1].legend()
    
    # Gráfico 3: Calinski-Harabasz Score
    if calinski_scores:
        axes[0, 2].plot(k_range, calinski_scores, 'mo-', linewidth=2, markersize=8)
        axes[0, 2].set_xlabel('Número de Clusters (k)', fontsize=12)
        axes[0, 2].set_ylabel('Calinski-Harabasz Score', fontsize=12)
        axes[0, 2].set_title('Método de Calinski-Harabasz', fontsize=14, fontweight='bold')
        axes[0, 2].grid(True, alpha=0.3)
        
        if 'calinski' in resultados:
            axes[0, 2].axvline(x=resultados['calinski'], color='red', linestyle='--',
                              label=f'Óptimo: k={resultados["calinski"]}')
            axes[0, 2].legend()
    
    # Gráfico 4: Davies-Bouldin Score
    if davies_scores:
        axes[1, 0].plot(k_range, davies_scores, 'co-', linewidth=2, markersize=8)
        axes[1, 0].set_xlabel('Número de Clusters (k)', fontsize=12)
        axes[1, 0].set_ylabel('Davies-Bouldin Score', fontsize=12)
        axes[1, 0].set_title('Método de Davies-Bouldin', fontsize=14, fontweight='bold')
        axes[1, 0].grid(True, alpha=0.3)
        
        if 'davies' in resultados:
            axes[1, 0].axvline(x=resultados['davies'], color='red', linestyle='--',
                              label=f'Óptimo: k={resultados["davies"]}')
            axes[1, 0].legend()
    
    # Gráfico 5: Ratio de reducción de inercia
    diff_ratio = np.array([(inertia[i-1] - inertia[i]) / inertia[i-1] 
                          for i in range(1, len(inertia))])
    axes[1, 1].plot(k_range[1:], diff_ratio, 'ro-', linewidth=2, markersize=8)
    axes[1, 1].set_xlabel('Número de Clusters (k)', fontsize=12)
    axes[1, 1].set_ylabel('Reducción Relativa de Inercia', fontsize=12)
    axes[1, 1].set_title('Ratio de Reducción de Inercia', fontsize=14, fontweight='bold')
    axes[1, 1].grid(True, alpha=0.3)
    axes[1, 1].axhline(y=0.05, color='gray', linestyle='--', alpha=0.5,
                      label='Umbral 5%')
    axes[1, 1].legend()
    
    # Gráfico 6: Resumen de métricas normalizadas
    axes[1, 2].axis('off')
    summary_text = "RESUMEN DE SUGERENCIAS:\n\n"
    
    for metodo_nombre, k_optimo in resultados.items():
        if metodo_nombre != 'sugerencia_final':
            summary_text += f"{metodo_nombre.capitalize()}: k = {k_optimo}\n"
    
    summary_text += f"\n{'='*30}\n"
    summary_text += f"SUGERENCIA FINAL: k = {sugerencia_final}"
    
    axes[1, 2].text(0.1, 0.5, summary_text, fontsize=12, 
                   verticalalignment='center', fontweight='bold')
    
    plt.suptitle(f'Análisis del Número Óptimo de Clusters\n'
                 f'Dataset: {len(X)} muestras, {X.shape[1]} features', 
                 fontsize=16, fontweight='bold')
    plt.tight_layout()
    plt.show()
    
    # Tabla resumen detallada
    print("\n" + "=" * 70)
    print("TABLA RESUMEN DE MÉTRICAS")
    print("=" * 70)
    
    resumen_df = pd.DataFrame({
        'k': list(k_range),
        'Inercia': inertia,
        'Silhouette': silhouette_scores if silhouette_scores else [0]*len(k_range),
        'Calinski-Harabasz': calinski_scores if calinski_scores else [0]*len(k_range),
        'Davies-Bouldin': davies_scores if davies_scores else [0]*len(k_range)
    })
    
    print(resumen_df.round(4).to_string())
    
    # Imprimir recomendaciones
    print("\n" + "=" * 70)
    print("RECOMENDACIONES FINALES")
    print("=" * 70)
    
    for metodo_nombre, k_optimo in resultados.items():
        if metodo_nombre != 'sugerencia_final':
            print(f"• {metodo_nombre.capitalize()}: {k_optimo} clusters")
    
    print(f"\nSUGERENCIA FINAL: {sugerencia_final} clusters")
    
    # Interpretación del silhouette score para la sugerencia final
    if 'silhouette' in resultados:
        k_silhouette = resultados['silhouette']
        idx = list(k_range).index(k_silhouette)
        score = silhouette_scores[idx]
        
        print(f"\nSilhouette Score para k={k_silhouette}: {score:.4f}")
        if score > 0.7:
            print("Excelente estructura de clusters")
        elif score > 0.5:
            print("Buena estructura de clusters")
        elif score > 0.25:
            print("Estructura débil pero aceptable")
        else:
            print("Estructura muy débil, reconsiderar clustering")
    
    return resultados, resumen_df

# Uso del código con tu dataframe
print("Analizando número óptimo de clusters para tu dataset...")
resultados, resumen_df = encontrar_numero_optimo_clusters(rfm_scaled, max_clusters=10, metodo='todos')
````

#### **1.2.3 Segmentos y Descuentos**

Ahora vamos a mapear el descuento a los clusters correspondientes y también para que se vea aplicado a cada cliente, de manera que el equipo de mercadeo sepa a cuál cliente específicamente aplicar cuál descuento.

````python
# Función robusta que maneja cualquier tipo de dato
def asignar_descuento(cluster_val):
    # Convertir a string para comparación segura
    cluster_str = str(cluster_val)
    
    if cluster_str == '0':
        return 0.00
    elif cluster_str == '1':
        return 0.25
    elif cluster_str == '2':
        return 0.20
    elif cluster_str == '3':
        return 0.05
    else:
        # Para cualquier otro valor de cluster no esperado
        return 0.00  # o podrías usar np.nan si prefieres

# Aplicar la función
df_perfiles['Descuento'] = df_perfiles['cluster'].apply(asignar_descuento)

# Convertir a float
df_perfiles['Descuento'] = df_perfiles['Descuento'].astype(float)

df_perfiles
````

````python
df_perfiles_agregado_median = df_perfiles.groupby('cluster').agg(
    Descuento=('Descuento', 'first'),  # Tomar el primer valor (todos son iguales por cluster)
    recency=('recency', 'median'),
    frequency=('frequency_total', 'median'),
    monetary=('monetary_total', 'median'),
    recency_compra=('compra_recency', 'median'),
    frequency_compra=('compra_frequency', 'median'),
    monetary_compra=('compra_monetary', 'median'),
    recency_avance=('avance_recency', 'median'),
    frequency_avance=('avance_frequency', 'median'),
    monetary_avance=('avance_monetary', 'median'),
    cluster_size=('monetary_total', 'count')
).round(2).sort_values(by='recency')

df_perfiles_agregado_median
````

#### **2.1 Cantidad de estudiantes por ciudad. Adicionalmente, ordenar estas ciudades por cantidad de estudiantes en ella de mayor a menor.**

```sql
SELECT c.NOMBRECIUDAD, COALESCE(COUNT(e.ID), 0) AS CantidadEstudiantes FROM Ciudad c LEFT JOIN Estudiante e ON e.CODIGOCIUDAD = c.CODIGOCIUDAD GROUP BY c.NOMBRECIUDAD ORDER BY CantidadEstudiantes DESC;
```

#### **2.2 Agrupar la cantidad de estudiantes por año de ingreso y carrera.**
```sql
SELECT YEAR(STR_TO_DATE(FECHAINGRESO, '%d/%m/%Y')) AS AnioIngreso, CARRERA, COUNT(ID) AS CantidadEstudiantes FROM Estudiante GROUP BY YEAR(STR_TO_DATE(FECHAINGRESO, '%d/%m/%Y')), CARRERA; 
```
En este caso estoy usando la función de MySQL ```STR_TO_DATE``` para asegurarme de que la fecha esté en el formato correcto.

También me percaté de, que hay algunos registros donde ```FECHAINGRESO``` es ```null```, entonces si lo que se quiere es que no aparezcan esos registros, podemos usar una sentencia ```WHERE FECHAINGRESO IS NOT NULL``` después de ```FROM Estudiante```, para que el resultado de la cosulta quede libre de ```null```. O bien, usar un ```COALESCE(YEAR(STR_TO_DATE(FECHAINGRESO, '%d/%m/%Y')), 0)``` para cambiar los años con ```null``` por 0.

#### **2.5 Seleccionar los estudiantes que no tienen Teléfono.**
```sql
SELECT NOMBREAPELLIDO FROM Estudiante WHERE TELEFONO IS NULL OR TELEFONO = ''; 
```

En este caso noté que en la base de datos hay números mal ingresados, tal que contienen letras o guiones bajos, o contienen otro tipo de errores. Entonces esos núumero también deberían ser considerados como que no existen, es decir, los estudiantes asociados a esos registros, no tienen teléfono, por lo tanto, una mejor consulta a prueba de estos problemas es la siguiente:

```sql
SELECT NOMBREAPELLIDO, TELEFONO FROM Estudiante WHERE TELEFONO IS NULL OR TELEFONO = '' OR TELEFONO REGEXP '[A-Za-z]' OR TELEFONO LIKE '%\_%' OR TELEFONO LIKE '\_%'; 
```

#### **2.7 Valide que el campo CORREO sea consistente, de lo contrario se debe marcar nulo. (Validar al menos que contenga el caracter '@').**

Como primera medida podríamos filtrar aquellos correos que contienen '@' como válidos, eso se logra de la siguiente manera:

```sql
SELECT ID, NOMBREAPELLIDO, CARRERA, CASE WHEN CORREO = '' OR CORREO NOT LIKE '%@%' THEN NULL ELSE CORREO END AS CorreoValidado FROM Estudiante; 
```
Sin embargo, esto no es suficiente para los registros en esta base de datos, ya que al visualizar ligeramente los correos, puedo notar que hay incluso correos que empiezan en '@' (esto no es válido ya que a pesar de que contenga '@', empezar con este caracter especial no es admitido), o correos que terminan en '.co' (si bien es cierto que un correo puede terminar en '.co', generalmente sucede cuando antes ya tenía otra extensión, por ejemplo, '.com.co' o '.edu.co', etc.). También hay correos que terminan en '.c', lo cual es un error. Por lo tanto, una solución más robusta para validar los correos, es la siguiente:

```sql
SELECT ID, NOMBREAPELLIDO, CARRERA, CASE WHEN CORREO = '' OR CORREO NOT LIKE '%@%' OR CORREO LIKE '@%' OR NOT (CORREO LIKE '%.com' OR CORREO REGEXP '\\.[A-Za-z]+\\.co$') THEN NULL ELSE CORREO END AS CorreoValidado FROM Estudiante;
```

Este código también garantiza que no se acepten correos como los que mencioné anteriormente, y que además estaban registrados en la base de datos.
